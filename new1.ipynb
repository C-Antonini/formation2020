# Fonction $\Gamma$

La fonction est définie pour tout $x>0$ par $\Gamma(x)=\displaystyle\int_0^{+\infty}t^{x-1}e^{-t}\,dt$

![au cas où](refraction.png)

```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
def Gamma(x):
    def f(t):
        return t**(x-1)*np.exp(-t)
    return quad(f,0,np.inf)[0]
X=[0.1+k/100 for k in range(401)]
Y=[Gamma(x) for x in X]
plt.plot(X,Y)
```
